```python
import keras

from common import *

from CNN.CNN_v0 import CNN_v0
```

```python
train, test = data_sets()
image_tensors = train[0], test[0]
person_coo_tensors = train[1], test[1]
count_matrix = train[2], test[2]

image_train, image_test = image_tensors
person_coo_train, person_coo_test = person_coo_tensors
count_train, count_test = count_matrix
count_train = count_train.astype(np.uint16)
count_test = count_test.astype(np.uint16)
```

```python
cnn_v0_40 = CNN_v0((480, 640, 3), split_into_parts=40)
cnn_v0_40.def_model()
```

```python
images = cnn_v0_40._prepare_images(image_train)
anwsers = cnn_v0_40._prepare_anwsers(person_coo_train)
```

```python
history = cnn_v0_40.model.fit(images[:1], anwsers[:1])
```

```python
parallel_model = keras.utils.multi_gpu_model(
        cnn_v0_40.model,
        gpus=6,
        cpu_merge=False)
parallel_model.compile(
        loss="binary_crossentropy", 
        optimizer="adam", 
        metrics=["accuracy"])
```

```python
parallel_model.fit(images, anwsers, batch_size=1177000, epochs=1,
                   validation_split=0.1)
```

```python
parallel_model.optimizer.lr.assign(1e-1)
```

```python
parallel_model.fit(images, anwsers, batch_size=1177000, epochs=100,
                   validation_split=0.1)
```
