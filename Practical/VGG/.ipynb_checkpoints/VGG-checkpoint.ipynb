# Importing VGG19

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import keras

mnist = tf.keras.datasets.mnist.load_data()

train_images = mnist[0][0]
train_anwsers = mnist[0][1]

train_images = train_images.reshape((-1, 28, 28, 1))
train_anwsers = keras.utils.np_utils.to_categorical(y=train_anwsers, num_classes=10)

test_images = mnist[1][0]
test_anwsers = mnist[1][1]

test_images = test_images.reshape((-1, 28, 28, 1))
test_anwsers = keras.utils.np_utils.to_categorical(y=test_anwsers, num_classes=10)
```

```python
model = keras.models.Sequential()

model.add(keras.layers.BatchNormalization(input_shape=(28, 28, 1)))
model.add(keras.layers.Conv2D(24, (3, 3), activation="relu"))

model.add(keras.layers.Flatten())

model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dense(10, activation="sigmoid"))

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(x=train_images, y=train_anwsers, batch_size=16, epochs=5)
```

```python
model.fit(x=train_images, y=train_anwsers, batch_size=300, epochs=5)
```

```python
model.fit(x=train_images, y=train_anwsers, batch_size=60000, epochs=5)
```

```python
history = model.fit(x=train_images, y=train_anwsers, batch_size=600, epochs=5)
```

```python
history = model.fit(x=train_images[:-1000], y=train_anwsers[:-1000],
		validation_data=(train_images[-1000:], train_anwsers[-1000:]), batch_size=300, epochs=5)
```

```python
model.optimizer.lr = 1e-5
```

```python
next_history = model.fit(x=train_images[:-1000], y=train_anwsers[:-1000],
		validation_data=(train_images[-1000:], train_anwsers[-1000:]), batch_size=300, epochs=5)
```

```python
test_history = model.evaluate(test_images, test_anwsers, batch_size=300)
```
